```python
#read data
data = {
    'tra': pd.read_csv('air_visit_data.csv'),
    'as': pd.read_csv('air_store_info_with_nearest_active_station.csv'),
    'hs': pd.read_csv('hpg_store_info_with_nearest_active_station.csv'),
    'ar': pd.read_csv('air_reserve.csv'),
    'hr': pd.read_csv('hpg_reserve.csv'),
    'id': pd.read_csv('store_id_relation.csv'),
    'tes': pd.read_csv('sample_submission.csv'),
    'hol': pd.read_csv('date_info.csv').rename(columns={'calendar_date':'visit_date'}),
    'wd': pd.read_csv('feature_manifest.csv').rename(columns={'id':'station_id'})
    }

data['hr'] = pd.merge(data['hr'], data['id'], how='inner', on=['hpg_store_id'])

#weather data
data['as']= data['as'].merge(wd,on='station_id',how='left')

#processing air_reserve and hpg_reserve (sum, mean values by visit_datetime)
for df in ['ar','hr']:
    data[df]['visit_datetime'] = pd.to_datetime(data[df]['visit_datetime']).dt.date
    data[df]['reserve_datetime'] = pd.to_datetime(data[df]['reserve_datetime']).dt.date
    data[df]['reserve_datetime_diff'] = data[df].apply(lambda r: (r['visit_datetime'] - r['reserve_datetime']).days, axis=1)
    tmp1 = data[df].groupby(['air_store_id','visit_datetime'], as_index=False)[['reserve_datetime_diff', 'reserve_visitors']].sum().rename(columns={'visit_datetime':'visit_date', 'reserve_datetime_diff': 'rs1', 'reserve_visitors':'rv1'})
    tmp2 = data[df].groupby(['air_store_id','visit_datetime'], as_index=False)[['reserve_datetime_diff', 'reserve_visitors']].mean().rename(columns={'visit_datetime':'visit_date', 'reserve_datetime_diff': 'rs2', 'reserve_visitors':'rv2'})
    data[df] = pd.merge(tmp1, tmp2, how='inner', on=['air_store_id','visit_date'])

#get datetime info for train (visit_data) and test
data['tra']['visit_date'] = pd.to_datetime(data['tra']['visit_date'])
data['tra']['dow'] = data['tra']['visit_date'].dt.dayofweek
data['tra']['year'] = data['tra']['visit_date'].dt.year
data['tra']['month'] = data['tra']['visit_date'].dt.month
data['tra']['visit_date'] = data['tra']['visit_date'].dt.date
data['tes']['visit_date'] = data['tes']['id'].map(lambda x: str(x).split('_')[2])
data['tes']['air_store_id'] = data['tes']['id'].map(lambda x: '_'.join(x.split('_')[:2]))
data['tes']['visit_date'] = pd.to_datetime(data['tes']['visit_date'])
data['tes']['dow'] = data['tes']['visit_date'].dt.dayofweek
data['tes']['year'] = data['tes']['visit_date'].dt.year
data['tes']['month'] = data['tes']['visit_date'].dt.month
data['tes']['visit_date'] = data['tes']['visit_date'].dt.date

#basic stats for restraunts per day of week 
unique_stores = data['tes']['air_store_id'].unique()
stores = pd.concat([pd.DataFrame({'air_store_id': unique_stores, 'dow': [i]*len(unique_stores)}) for i in range(7)], axis=0, ignore_index=True).reset_index(drop=True)
tmp = data['tra'].groupby(['air_store_id','dow']).agg({'visitors' : [np.min,np.mean,np.median,np.max,np.size]}).reset_index()
tmp.columns = ['air_store_id', 'dow', 'min_visitors', 'mean_visitors', 'median_visitors','max_visitors','count_observations']
stores = pd.merge(stores, tmp, how='left', on=['air_store_id','dow']) 
stores = pd.merge(stores, data['as'], how='left', on=['air_store_id']) 

#label encoding for genre/area restraunt name 
stores['air_genre_name'] = stores['air_genre_name'].map(lambda x: str(str(x).replace('/',' ')))
stores['air_area_name'] = stores['air_area_name'].map(lambda x: str(str(x).replace('-',' ')))
lbl = preprocessing.LabelEncoder()
for i in range(10):
    stores['air_genre_name'+str(i)] = lbl.fit_transform(stores['air_genre_name'].map(lambda x: str(str(x).split(' ')[i]) if len(str(x).split(' '))>i else ''))
    stores['air_area_name'+str(i)] = lbl.fit_transform(stores['air_area_name'].map(lambda x: str(str(x).split(' ')[i]) if len(str(x).split(' '))>i else ''))
stores['air_genre_name'] = lbl.fit_transform(stores['air_genre_name'])
stores['air_area_name'] = lbl.fit_transform(stores['air_area_name'])

#label encoding for day_of_week restraunt name 
data['hol']['visit_date'] = pd.to_datetime(data['hol']['visit_date'])
data['hol']['day_of_week'] = lbl.fit_transform(data['hol']['day_of_week'])
data['hol']['visit_date'] = data['hol']['visit_date'].dt.date

#merge encoded day_of_week with train/test data 
train = pd.merge(data['tra'], data['hol'], how='left', on=['visit_date']) 
test = pd.merge(data['tes'], data['hol'], how='left', on=['visit_date']) 

#merge encoded stores names with train/test data 
train = pd.merge(train, stores, how='left', on=['air_store_id','dow']) 
test = pd.merge(test, stores, how='left', on=['air_store_id','dow'])

#merge air/hpg reserve with train/test data 
for df in ['ar','hr']:
    train = pd.merge(train, data[df], how='left', on=['air_store_id','visit_date']) 
    test = pd.merge(test, data[df], how='left', on=['air_store_id','visit_date'])

train['id'] = train.apply(lambda r: '_'.join([str(r['air_store_id']), str(r['visit_date'])]), axis=1)

#additional stats
train['total_reserv_sum'] = train['rv1_x'] + train['rv1_y']
train['total_reserv_mean'] = (train['rv2_x'] + train['rv2_y']) / 2
train['total_reserv_dt_diff_mean'] = (train['rs2_x'] + train['rs2_y']) / 2

test['total_reserv_sum'] = test['rv1_x'] + test['rv1_y']
test['total_reserv_mean'] = (test['rv2_x'] + test['rv2_y']) / 2
test['total_reserv_dt_diff_mean'] = (test['rs2_x'] + test['rs2_y']) / 2
```

```python
#additional stats 2
train['date_int'] = train['visit_date'].apply(lambda x: x.strftime('%Y%m%d')).astype(int)
test['date_int'] = test['visit_date'].apply(lambda x: x.strftime('%Y%m%d')).astype(int)
train['var_max_lat'] = train['latitude'].max() - train['latitude']
train['var_max_long'] = train['longitude'].max() - train['longitude']
test['var_max_lat'] = test['latitude'].max() - test['latitude']
test['var_max_long'] = test['longitude'].max() - test['longitude']

train['lon_plus_lat'] = train['longitude'] + train['latitude'] 
test['lon_plus_lat'] = test['longitude'] + test['latitude']

lbl = preprocessing.LabelEncoder()
train['air_store_id2'] = lbl.fit_transform(train['air_store_id'])
test['air_store_id2'] = lbl.transform(test['air_store_id'])
```

```python
#removing outliers
outliers = (train.groupby( ['air_store_id'])['visitors'].transform(zscore))
train.drop(outliers[outliers>3].index,inplace=True)
print(str(len(outliers[outliers>3].index))+' outliers have been removed')
```

```python
col = [c for c in train if c not in ['id', 'air_store_id', 'visit_date','visitors','station_id','latitude_str', 'longitude_str']]
train = train.fillna(-1)
test = test.fillna(-1)
```

```python
def RMSLE(y, pred):
    return metrics.mean_squared_error(y, pred)**0.5
```

```python
#17:12
# model1 = ensemble.GradientBoostingRegressor(learning_rate=0.2, random_state=123, n_estimators=200, subsample=0.8, max_depth =10,verbose=True)
# model1.fit(train[col], np.log1p(train['visitors'].values))
Xgbmodel= XGBRegressor(learning_rate=0.2, random_state=123, n_estimators=200, subsample=0.8, colsample_bytree=0.8, max_depth =10,n_jobs=-1)
# Xgbmodel.fit(train[col], np.log1p(train['visitors'].values),verbose=True,eval_metric='rmse')
```

```python
%%time
#time-series validation
tscv = TimeSeriesSplit(n_splits=10)
res_13_01=[]
for train_index, test_index in tscv.split(train):
    Xgbmodel.fit(train.iloc[train_index].loc[:,col], np.log1p(train.iloc[train_index].loc[:,'visitors'].values),verbose=True)
    res_13_01.append(RMSLE(np.log1p(train.iloc[test_index].loc[:,'visitors'].values), Xgbmodel.predict(train.iloc[test_index].loc[:,col])))
```

```python
#feature importance
plot_importance(Xgbmodel)
thresholds = np.sort(Xgbmodel.feature_importances_)
for thresh in thresholds:
    # select features using threshold
    selection = SelectFromModel(Xgbmodel, threshold=thresh, prefit=True)
    select_X_train = selection.transform(train.iloc[train_index].loc[:,col])
    # train model
    selection_model = XGBRegressor()
    selection_model.fit(select_X_train, np.log1p(train.iloc[train_index].loc[:,'visitors'].values))
    # eval model
    select_X_test = selection.transform(train.iloc[test_index].loc[:,col])
    y_pred = selection_model.predict(select_X_test)
    # predictions = [round(value) for value in y_pred]
    accuracy = RMSLE(np.log1p(train.iloc[test_index].loc[:,'visitors'].values), y_pred)
    print("Thresh=%.3f, n=%d, RMSLE: %.2f%" % (thresh, select_X_train.shape[1], accuracy*100.0))
```
